```python
import pandas as pd
from scipy import sparse
from sklearn.model_selection import train_test_split
```

```python
ratings_ = pd.read_csv('data/rating.csv')
movies_ = pd.read_csv('data/movie.csv')
ratings_merged= pd.merge(movies_,ratings_).drop(['genres','timestamp'],axis=1)
print(ratings_.shape)
ratings_.head()
```

```python
print(ratings_merged.shape)
ratings_merged.head()
```

```python
grouped = ratings_.groupby('movieId')['rating'].count()
df_new = grouped.reset_index()
dff = df_new[df_new['rating'] >= 50]
ratings = pd.merge(ratings_merged, dff, left_on='movieId', right_on='movieId', how='inner').drop('rating_y', axis=1) 
ratings
```
